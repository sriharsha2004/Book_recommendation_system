# **Popularity Based recommendation System**

```python
# Getting the top popular books out of all books

# For getting popular books , we find books with more than 250 reviews and also
# the average rating for the book is above certain threshold

popular_df = books.merge(ratings,on="ISBN")
# merging books with ratings , to get access to ratings of every book

popular_df
```

```python
num_rating_df = popular_df.groupby("Book-Title").count()["Book-Rating"].reset_index()
num_rating_df.rename(columns={"Book-Rating" : "num_ratings"},inplace=True)
num_rating_df
# dataframe having => books - no of ratings or review
```

```python
avg_rating_df = popular_df.groupby('Book-Title').agg({'Book-Rating':[np.mean]}).reset_index()

# Flatten multi-level column index
avg_rating_df.columns = avg_rating_df.columns.get_level_values(0)

avg_rating_df.rename(columns={'Book-Rating': 'avg_rating'},inplace=True)
avg_rating_df
# datafrom having => books - average of ratings
```

```python
popular_books_df = num_rating_df.merge(avg_rating_df , on="Book-Title")
```

```python
popular_books_df
```

```python
# .boolean indexing

# All records with ratings greater than 250
popular_books_df = popular_books_df[popular_books_df["num_ratings"] > 250]

# Sorting in descending on avg_rating and then extract first 50
popular_books_df = popular_books_df.sort_values('avg_rating',ascending=False).head(50)

# This gives us top 50 popular books
popular_books_df
```

```python
# We also need thee book-title , image of book to display

popular_books_df = popular_books_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
popular_books_df
```

```python
import pickle
pickle.dump(popular_books_df,open('popular.pkl','wb'))
```

# **Collaborative Filtering Based Recommender System**

```python
# Recommendation system based on user search

# We will build a df => book1 user1rating user2rating user3rating user4rating .....
                    # book2 ....
                    # book3 ....
                    # book 196 ...

# Now we can treat the user ratings of each book as vector and find cosine similarity

# We will only consider the users who gave more than 200 reviews --> pada likha user only
# And only books having more than 200 reviews
```

```python
# boolean indexing
x = popular_df.groupby('User-ID').count()['Book-Rating'] > 200
# getting no of reviews given by each user (getting users with more than 200 only)

book_readers = x[x].index
# getting index of every record extracted i.e index of every user who gave more than 200 reviews
```

```python
book_readers
# indexes of users
```

```python
filtered_rating = popular_df[popular_df['User-ID'].isin(book_readers)]
# getting data of all those users
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
# filtering all books with more than 50 reviews

famous_books = y[y].index
# getting thier indexes
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
# creating a table using pivot_table
```

```python
pt.head()
```

```python
pt.fillna(0,inplace=True)
# Fill Nan with 0
```

```python
pt.head()
```

```python
# Using a neighrest neighbours algorithm which is a clustering algorithm
# It internally calculates distances between all the vectors and forms cluster

# But it has several zeros , so we have to make sure that distances are not calculated between zeros
# So we are creating book_sparse

from scipy.sparse import csr_matrix
book_sparse = csr_matrix(pt)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm="brute")
# Using brute indicates find distances between all
```

```python
model.fit(book_sparse)
```

```python
# We have to use a function kneighbors where we have to define the input with which distance has to be found
# with others
# and the number of neighbors to extract

# getting distances from pt[237] to others
# extracting 6 nearest neighbours

distances , suggestions = model.kneighbors(pt.iloc[237,:].values.reshape(1,-1),n_neighbors=6);

# It returns two arrays :
# first array returns the distances of books from the specified book i.e with index 237 i.e nearest
# second array returns the indexes of those shorter distances
```

```python
distances
# first distance is 0 i.e with itself
```

```python
suggestions
# 237 is with itself
```

```python
# gettings names of those books
for i in range(len(suggestions)):
  print(pt.index[suggestions[i]])
```

```python
# function to get related books
def recommend_book(book_name):
  book_id = np.where(pt.index==book_name)[0][0]
  # print(book_id)
  distances , suggestions = model.kneighbors(pt.iloc[book_id,:].values.reshape(1,-1),n_neighbors=6);
  res = []
  for i in range(len(suggestions[0])):
    item = []
    if(i!=0):
      # prints the name of books
      # print(pt.index[suggestions[0][i]])

      # gettings details of that book
      temp_df = books[books["Book-Title"] == pt.index[suggestions[0][i]]]
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
      item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

      res.append(item)

  return res
```

# **We can also use cosine similarity**

```python
# Findnig cosine sililarity from one book to other book using the vecctors [user1,user2,user3,...]

from sklearn.metrics.pairwise import cosine_similarity

similarity_scores = cosine_similarity(pt)
# finding similaity of one with all in pivaot table
```

```python
# function which takes the book name and recommends the books

def recommend(book_name):

    index = np.where(pt.index==book_name)[0][0]

  # finding similarity score of that with other books
  # enumerate converts values into (indexes,values)
  # Sort based on the similarity scores descending and get first 5
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        # removing duplicate books
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend("A Bend in the Road")
```
